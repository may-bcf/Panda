```python
df2010r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2010, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2010, 1, 1))].revenue.mean()
df2011r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2011, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2011, 1, 1))].revenue.mean() 
df2012r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2012, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2012, 1, 1))].revenue.mean() 
df2013r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2013, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2013, 1, 1))].revenue.mean() 
df2014r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2014, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2014, 1, 1))].revenue.mean() 
df2015r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2015, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2015, 1, 1))].revenue.mean() 
df2016r = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2016, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2016, 1, 1))].revenue.mean() 
```

```python
df2010b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2010, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2010, 1, 1))].budget.mean()
df2011b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2011, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2011, 1, 1))].budget.mean() 
df2012b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2012, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2012, 1, 1))].budget.mean() 
df2013b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2013, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2013, 1, 1))].budget.mean() 
df2014b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2014, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2014, 1, 1))].budget.mean() 
df2015b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2015, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2015, 1, 1))].budget.mean() 
df2016b = df5[(pd.to_datetime(df1['release_date']) < dt.datetime(2016, 12, 31)) &       
          (pd.to_datetime(df1['release_date']) > dt.datetime(2016, 1, 1))].budget.mean()
```

```python
fig, ax = plt.subplots()

years = np.array(['2010','2011','2012','2013','2014','2015','2016'])
revenues = [df2010r, df2011r, df2012r, df2013r,df2014r, df2015r, df2016r ]

ax.bar(years, revenues)
```

```python
def million(x, pos):
        return 'PLN {:2.1f}M'.format(x*1e-6)
```

```python
fig, ax = plt.subplots()

years = np.array(['2010','2011','2012','2013','2014','2015','2016'])
revenues = [df2010r, df2011r, df2012r, df2013r,df2014r, df2015r, df2016r ]
budget = [df2010b, df2011b, df2012b, df2013b,df2014b, df2015b, df2016b ]

formatter = plt.FuncFormatter(million)

ax.yaxis.set_major_formatter(formatter)
ax.plot(budget,color='red', linewidth=3, label = 'budget')
ax.bar(years, revenues,label = 'przychód')

ax.legend(loc=(1.05,0.8))

ax.set_title('Średni przychód i budżet filmów w latach 2010 - 2016')
```
