```python
df = pd.read_csv( 'calibration.log',header=None ,on_bad_lines='skip')
df2 = df[0]
prefixes = ['2022-06-18', '2022-06-19', '2022-06-20' ]
df3 = df2[df2.str.startswith(tuple(prefixes))]
df4 = df3.str.split(' ', expand=True)
df4.columns = ["date", "time", "2", "3",  "4",  "5", "stored" , "7" , "mozaic" , "9", "10", "11" ,"12", "fcuk", "14" ]
df4['date'] =df4["date"] + " " + df4['time']
df8 = df4[(df4['stored'] == 'Mosaic')]
df9 = df8.drop_duplicates(subset=['fcuk'])
cal_rest = df9[['date', 'mozaic', 'fcuk']]
cal_rest
```

```python
df = pd.read_csv( 'gateway.log',header=None ,on_bad_lines='skip')
df2 = df[0]
prefixes = ['2022-06-18', '2022-06-19', '2022-06-20' ]

df3 = df2[df2.str.startswith(tuple(prefixes))]
df4 = df3.str.split(' ', expand=True)
df4.columns = ["date", "time", "2", "3",  "4",  "5", "stored" , "7" , "8" , "mozaic", "10", "11" ,"12", "13", "14" ]
df4['date'] =df4["date"] + " " + df4['time']
gateway = df4[(df4['stored'] == 'Stored') & (df4['mozaic'] != 'from')]
gateway['dt'] = pd.to_datetime(gateway['date'])
gateway


```

```python
# mozaik 104 check
```

```python
df104 = gateway[(gateway['mozaic'] == '104.')]
df104 = df104[['date', 'time', 'mozaic']]
df104
```

```python
df104['dt'] = pd.to_datetime(df104['date'])
df104['date_diff'] = df104['dt'].diff()
```

```python
df104.describe()
```

```python
df104check = df104[(df104['date_diff'] > '0 days 00:02:00.00')]
```

```python
df104check
```

```python
df104check.to_csv('check104.csv')
```
